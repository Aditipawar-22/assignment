```python
#Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
#plot a scatter plot.
import matplotlib.pyplot as p
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
p.scatter(x,y)
p.title("Scatter diagram of x & y :")
p.xlabel("X values ")
p.ylabel("Y values ")
#Note: Also add title, xlabel, and ylabel to the plot.
```

```python
#Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
#Use the following data:

import matplotlib.pyplot as plt
import numpy as np

x = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])



plt.subplot(2, 2, 1) 
plt.plot(x, y1, label='Line 1')
plt.legend()

plt.subplot(2, 2, 2)  
plt.plot(x, y2, label='Line 2')
plt.legend()

plt.subplot(2, 2, 3)  
plt.plot(x, y3, label='Line 3')
plt.legend()

plt.plot(x, y4, label='Line 4')
plt.legend()

plt.tight_layout()  

plt.show()
```

```python
'''The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It is
particularly useful when you want to display multiple plots in a single figure for easy comparison. The function
takes three arguments: the number of rows, the number of columns, and the index of the subplot you want to create.'''
```

```python
'''Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.'''
import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.barh(company,profit)
plt.show()
```

```python
#Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
'''A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a
dataset. It displays the summary of a set of data points by showing the median, quartiles, and potential outliers.
The box in the plot represents the interquartile range (IQR), which spans from the first quartile (Q1) to the third
quartile (Q3), with a line indicating the median. The "whiskers" extend from the box to the minimum and maximum
values within a certain range (usually 1.5 times the IQR), and any data points outside this range are considered
outliers.
'''

import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.boxplot(data, labels=['Box 1', 'Box 2'])
plt.title('Box Plot Example')
plt.ylabel('Value')
plt.xlabel('Boxes')

plt.show()
```
